```python
from torch_peaks.gaussians import Gaussian2D, WarpedGaussian2D
import matplotlib.pyplot as plt
import torch

g1 = Gaussian2D(1.,0., 0., 1., 1.)

xx, yy = torch.meshgrid(
    torch.linspace(-5, 5, 100),
    torch.linspace(-5, 5, 100),
    indexing='ij'
)

im_g1 = g1(xx,yy)
```

```python
plt.imshow(im_g1.detach().numpy(), cmap='gray')
```

```python
g2 = WarpedGaussian2D(1.,0.,0., 1.0, 1.0, 0.2, 0.)
```

```python
im_g2 = g2(xx,yy)
plt.imshow(im_g2.detach().numpy(), cmap='gray')
```
